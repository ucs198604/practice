### let the program guess your number
1. devise a positive number from 0000 to 9999, without repeated components (for example, 1234 and 4291 are legit; 0122 and 9242 are not legit due to repeated components)
2. if one component is correct in both digit and position → added in 'A' category
3. if one component is correct in digit but incorrect in position → added in 'B' category
4. reply the counts in AB category
> for example, if your number is `0123`, and the program guesses `0937`, it's `1A`(for 0) `1B`(for 3). You reply 11(for 1A1B)
    
Press Q or q to quit

```python
import random

# check ?A?B of 2 numbers, return ?A?B
def check_AB(num1, num2):
    A=0
    B=0
    for i in range(0,4):
        if num1[i]==num2[i]:
            A=A+1
        elif num1[i] in num2:
            B=B+1
    A = str(A)
    B = str(B)
    return (A,B)

# to delete the numbers that don't match current A/B test result and return
def clean(cbin_in,num,A,B):
    to_del=[]
    ini_num= len(cbin_in)
    cbin_out = cbin_in
    for i in range(0,len(cbin_in)):
        cbin_in[i] = str(cbin_in[i])
        if (check_AB(cbin_in[i],num)[0] != A) or (check_AB(cbin_in[i],num)[1] != B):
            to_del.append(cbin_in[i])
    for item in to_del:
        cbin_out.remove(item)
    end_num= len(cbin_out)
    print('total items:',end_num,'\ntoal eliminated:',ini_num-end_num)
    return cbin_out

# to create initial number list that has no repeated elements
def init_bin():
    # initiate bin, without repitition
    cbin0 = list(range(1,10000))
    repeat = []
    cguess=[]   # nx3 array for number/A/B
    for i in range(0,9999): # to make it as string
        cbin0[i]=str(cbin0[i])
        if len(cbin0[i])==1:
            cbin0[i]='000'+cbin0[i]
        elif len(cbin0[i])==2:
            cbin0[i]='00'+cbin0[i]        
        elif len(cbin0[i])==3:
            cbin0[i]='0'+cbin0[i]
        if len(set(cbin0[i]))!=4:   # if repeat, then set will shrink
            repeat.append(cbin0[i])
    for item in repeat:   # remove numbers that has repeated number
        cbin0.remove(item)
    return cbin0


# In[2]:



running=True
cbin = init_bin()
guess_list = []  # store gussed and result
print('''
# let the program guess your number
1. devise a positive number from 0000 to 9999, without repeated components (for example, 1234 and 4291 are legit; 0122 and 9242 are not legit due to repeated components)
2. if one component is correct in both digit and position → added in 'A' category
3. if one component is correct in digit but incorrect in position → added in 'B' category
4. reply the counts in AB category
> for example, if your number is `0123`, and the program guesses `0937`, it's `1A`(for 0) `1B`(for 3). You reply 11(for 1A1B)
    
Press Q or q to quit

''')

while running:
    current_num = random.choice(cbin)
    get_input= True
    while get_input:
        current_AB = input(f'Is it {current_num}, ?A?B? ')
        if (current_AB == 'q') or (current_AB == 'Q'):
            get_input = False
            print('quitted')
            break
        elif len(current_AB)!=2:
            print('Input must = 2')
        elif (current_AB[0] not in ['0','1','2','3','4']) or (current_AB[1] not in ['0','1','2','3','4']):
            print('Input false')
            
        elif int(current_AB[0])+ int(current_AB[1]) not in [0,1,2,3,4]:
            print('Input should between 0-4')
        else:
            get_input = False
            guess_list.append([current_num,current_AB[0], current_AB[1]])
    if (current_AB == 'q') or (current_AB == 'Q'):
        break
    
    # print current guess list
    for i in range(0,len(guess_list)):
        print(f'No.{i+1} guess, {guess_list[i][0]}, {guess_list[i][1]}A{guess_list[i][2]}B')
    
    # clear the list that doesn't match
    cbin = clean(cbin,current_num,current_AB[0], current_AB[1])
    print('Candidates:\n',cbin)
    
    if current_AB[0]=='4':
        running=False
        print(f'The answer is {current_num}, it takes {len(guess_list)+1} steps in total.')
    if not cbin:
        print('Oops, there must be something wrong...')
        running=False
        
print('======== end of game =======')

```
