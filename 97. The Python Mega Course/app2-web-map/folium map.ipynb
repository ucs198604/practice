根據不同高度，有不同的標籤顏色

```python
import pandas as pd
voc = pd.read_csv('Volcanoes.txt')

lat = voc['LAT'].tolist()
long = voc['LON'].tolist()
elev = voc['ELEV'].tolist()

import folium
m = folium.Map()
fg = folium.FeatureGroup(name='Group1')

def elevation_color(elevation):
    if elevation<=1000:
        return 'green'
    elif 1000<elevation<=2000:
        return 'blue'
    elif 2000<elevation<=3000:
        return 'orange'
    elif elevation>3000:
        return 'red'

for lat_, long_, elev_ in zip(lat,long,elev):
    child = folium.CircleMarker(location=(lat_,long_),
                         color=elevation_color(elev_), 
                        fill=True,
                         opacity =0.8)
    fg.add_child(child=child)

    
m.add_child(fg)

```
